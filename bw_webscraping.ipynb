```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
base_url = "https://www.kammi.it/scarpe-donna/"
sneakers = "sneakers/"
sandali_bassi = "sandali-bassi/"
stivali = "stivali-stivaletti/"
zeppe = "zeppe/"
sandali_tacco = "sandali-con-tacco/"
scarpe_tacco = "scarpe-con-tacco/"
scarpe_basse = "scarpe-basse/"
```

## Nome

```python
page_sneakers = requests.get(base_url+sneakers)
page_sandali_bassi = requests.get(base_url+sandali_bassi)
page_stivali = requests.get(base_url+stivali)
page_zeppe = requests.get(base_url+zeppe)
page_sandali_tacco = requests.get(base_url+sandali_tacco)
page_scarpe_tacco = requests.get(base_url+scarpe_tacco)
page_scarpe_basse = requests.get(base_url+scarpe_basse)
```

```python
#soup

soup_sneakers = BeautifulSoup(page_sneakers.text)
soup_sandali_bassi = BeautifulSoup(page_sandali_bassi.text)
soup_stivali = BeautifulSoup(page_stivali.text)
soup_zeppe = BeautifulSoup(page_zeppe.text)
soup_sandali_tacco = BeautifulSoup(page_sandali_tacco.text)
soup_scarpe_tacco = BeautifulSoup(page_scarpe_tacco.text)
soup_scarpe_basse = BeautifulSoup(page_scarpe_basse.text)
```

```python
sneakers_raw = soup_sneakers.find_all("p", attrs={"class": "product-title"})
sandali_bassi_raw = soup_sandali_bassi.find_all("p", attrs={"class": "product-title"})
stivali_raw = soup_stivali.find_all("p", attrs={"class": "product-title"})
zeppe_raw = soup_zeppe.find_all("p", attrs={"class": "product-title"})
sandali_tacco_raw = soup_sandali_tacco.find_all("p", attrs={"class": "product-title"})
scarpe_tacco_raw = soup_scarpe_tacco.find_all("p", attrs={"class": "product-title"})
scarpe_basse_raw = soup_scarpe_basse.find_all("p", attrs={"class": "product-title"})
```

```python
sneakers_name = []
for item in sneakers_raw:
    sneakers_name.append(item.text.strip())

    
sandali_bassi_name = []
for item in sandali_bassi_raw:
    sandali_bassi_name.append(item.text.strip())

stivali_name = []
for item in stivali_raw:
    stivali_name.append(item.text.strip())

zeppe_name = []
for item in zeppe_raw:
    zeppe_name.append(item.text.strip())

sandali_tacco_name = []
for item in sandali_tacco_raw:
    sandali_tacco_name.append(item.text.strip())

scarpe_tacco_name = []
for item in scarpe_tacco_raw:
    scarpe_tacco_name.append(item.text.strip())

scarpe_basse_name = []
for item in scarpe_basse_raw:
    scarpe_basse_name.append(item.text.strip())
```
